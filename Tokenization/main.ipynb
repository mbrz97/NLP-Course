```python
import nltk
from nltk.tokenize import WhitespaceTokenizer
```

```python
# Deleting spaces and tabs

input_filenames = ['sport.txt', 'tech.txt']
output_filenames = ['sport1.txt', 'tech1.txt']

for i in range(len(input_filenames)):
    with open(input_filenames[i], "r") as f:
        input_text = f.read()

    # Tokenize the text using WhiteSpaceTokenizer
    tokenizer = WhitespaceTokenizer()
    tokens = tokenizer.tokenize(input_text)
    output_text = ' '.join(tokens)

    # Save the tokens to an output text file
    with open(output_filenames[i], "w") as f:
        f.write(output_text)
```

```python
# Delete all characters except alphabet letters

from nltk.tokenize import RegexpTokenizer

input_filenames = ['sport1.txt', 'tech1.txt']
output_filenames = ['sport2.txt', 'tech2.txt']

tokenizer = RegexpTokenizer(r'[a-zA-Z]+')

for i in range(len(input_filenames)):
    with open(input_filenames[i], "r") as f:
        input_text = f.read()
        words = tokenizer.tokenize(input_text.lower())

        # Save the tokens to an output text file
        with open(output_filenames[i], "w") as f_out:
            f_out.write(" ".join(words))
```

```python
input_filenames = ['sport2.txt', 'tech2.txt']
output_filenames = ['sport3.txt', 'tech3.txt']
for i in range(len(input_filenames)):
     with open(input_filenames[i], 'r') as file:
         text = file.read()

     # Split the text into words
     words = text.split()

     # Remove one-character words
     words = [word for word in words if len(word) > 1]

     # Join the remaining words back into a string
     updated_text = ' '.join(words)

     # Write the updated text back into the file
     with open(output_filenames[i], 'w') as file:
         file.write(updated_text)
```

```python
# Removing stopwords

nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

input_filenames = ['sport3.txt', 'tech3.txt']
output_filenames = ['sport4.txt', 'tech4.txt']

stop_words = set(stopwords.words('english'))

for i in range(len(input_filenames)):
    with open(input_filenames[i], 'r') as f:
        text = f.read()
        words = word_tokenize(text)
        cleaned_words = [word for word in words if word not in stop_words]
        cleaned_text = ' '.join(cleaned_words)

        with open(output_filenames[i], 'w') as f_out:
            f_out.write(cleaned_text)

        print(input_filenames[i], 'Before:', len(text), 'After:', len(cleaned_text))
```

```python
input_filenames = ['sport4.txt', 'tech4.txt']
football_count = {'sport': 0, 'tech': 0}
computer_count = {'sport': 0, 'tech': 0}

for i in range(len(input_filenames)):
    with open(input_filenames[i], 'r') as file:
        text = file.read()

    # Convert the text to lowercase for case-insensitive matching
    text = text.lower()

    # Count the occurrences of the words
    football_count['sport'] = text.count('football')
    computer_count['tech'] = text.count('computer')


print("Occurrences of 'football':", football_count)
print("Occurrences of 'computer':", computer_count)
```

```python
from collections import Counter
nltk.download('averaged_perceptron_tagger')

def count_word_types(filename):
    with open(filename, 'r') as file:
        text = file.read()
        # Tokenize the text
        tokens = nltk.word_tokenize(text)
        # Perform POS tagging
        tagged_words = nltk.pos_tag(tokens)
        # Count the occurrences of each word type
        word_types = [tag[1] for tag in tagged_words]
        word_type_counts = Counter(word_types)
        print("---------- FILE NAME: ", filename,"----------")
        for word_type, count in word_type_counts.items():
            print(f'{word_type}: {count}')
        print("---------- ----------")
        return str(filename), word_type_counts
# Count the word types in the text file
word_type_counts1 = count_word_types('sport4.txt')
word_type_counts2 = count_word_types('tech4.txt')

with open('word_type.txt', 'w') as file:
    file.write(str(word_type_counts1))
    file.write('\n')
    file.write(str(word_type_counts2))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np

def create_word_term_matrix(file1_path, file2_path):
    # Read the contents of the first file
    with open(file1_path, 'r') as file1:
        text1 = file1.read()

    # Read the contents of the second file
    with open(file2_path, 'r') as file2:
        text2 = file2.read()

    # Combine the text from both files into a single list
    combined_text = [text1, text2]

    # Create an instance of CountVectorizer
    vectorizer = CountVectorizer()

    # Fit the vectorizer to the combined text and transform it into a word-term matrix
    word_term_matrix = vectorizer.fit_transform(combined_text)

    # Convert the matrix to a NumPy array
    word_term_matrix = word_term_matrix.toarray()

    # Get the feature names (words) from the vectorizer's vocabulary
    words = vectorizer.get_feature_names_out()
    return word_term_matrix, words

file1_path = 'sport4.txt'
file2_path = 'tech4.txt'
matrix, words = create_word_term_matrix(file1_path, file2_path)

# Print the word-term matrix
print("Word-Term Matrix:")

with open('word_term_matrix.txt', 'w') as file:
    for i, word in enumerate(words):
        print(f"{word}: {matrix[:, i]}")
        file.write(f"{word}: {matrix[:, i]} \n")
# first element is word count in sport.xt and second element is count in tech.txt
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def calculate_cosine_similarity(file1_path, file2_path, word1, word2):
    with open(file1_path, 'r') as file1:
        text1 = file1.read()
    with open(file2_path, 'r') as file2:
        text2 = file2.read()
    # Combine the text from both files into a list
    text_list = [text1.lower(), text2.lower()]
    # Create an instance of CountVectorizer
    vectorizer = CountVectorizer()
    # Fit the vectorizer to the text and transform it into a word-term matrix
    word_term_matrix = vectorizer.fit_transform(text_list)
    # Convert the word-term matrix to an array
    word_term_array = word_term_matrix.toarray()
    # Get the index of the words in the feature names (vocabulary)
    word1_index = vectorizer.vocabulary_.get(word1)
    word2_index = vectorizer.vocabulary_.get(word2)
    if word1_index is None or word2_index is None:
        return "One or both words not found in the vocabulary."
    # Get the vectors representing the occurrences of the words in each file
    word1_vector = word_term_array[:, word1_index]
    word2_vector = word_term_array[:, word2_index]
    # Reshape the vectors to match the expected input shape for cosine_similarity
    word1_vector = word1_vector.reshape(1, -1)
    word2_vector = word2_vector.reshape(1, -1)
    # Calculate the cosine similarity between the word vectors
    similarity = cosine_similarity(word1_vector, word2_vector)[0, 0]
    return similarity

file1_path = 'tech4.txt'
file2_path = 'sport4.txt'

ref_word = {'football':{'sport': 0, 'technology': 0, 'computer': 0},
            'sport': {'computer': 0, 'technology': 0, 'basketball': 0},
            'computer': {'basketball': 0, 'technology': 0, 'laptop': 0},
            'website': {'laptop': 0, 'technology': 0, 'football':0}
            }

for key, value in ref_word.items():
    for word in value:
        similarity = calculate_cosine_similarity(file1_path, file2_path, key, word)
        value[word] = similarity

with open('cosine_similarity.txt', 'w') as file:
    for key, value in ref_word.items():
        print(f"{key} -> {value}")
        print()  # Print a line break after each item
        file.write(f"{key} -> {value}\n")
```
